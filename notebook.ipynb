```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Filter the DataFrame for movies released in the 1990s
movies_90s = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]

# Select the duration column and find the most frequent movie duration
most_frequent_duration = movies_90s['duration'].mode()[0]

# Save the most frequent duration into a variable named "duration"
duration = most_frequent_duration

#Short Action Movies
short_action_movies = movies_90s["genre"]

# Filter movies with duration less than 90 minutes
short_movies = movies_90s[(movies_90s['genre'] == "Action") & (movies_90s['duration'] < 90)]
print(short_movies)
# Count the number of short movies
short_movie_count = short_movies.shape[0]

print(short_movie_count)
```
